```python
import numpy as np
import matplotlib.pyplot as plt

theta_axis = np.arange(0, 1, 0.001)

pxd1 = np.array([105*(x**2)*(1-x)**4 for x in theta_axis])
pxd5 = np.array([2772*(x**5)*(1-x)**5 for x in theta_axis])
pxd10 = np.array([102960*(x**8)*(1-x)**7 for x in theta_axis])

plt.plot(theta_axis, pxd1, color='red')
plt.plot(theta_axis, pxd5, color='blue')
plt.plot(theta_axis, pxd10, color='green')
plt.show()
```
